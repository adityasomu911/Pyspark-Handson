```python
#This sheet contains codes which is solving the problems given below.
#This is coded on Databricks python notebook on Pyspark Server, thus no sparkContext is created
"""Data source: dbfs:/databricks-datasets/adult/adult.data
Data Snapshot:
 
1.	Create a table of name – ‘Adult’ and load the data into the table to perform analysis.
2.	Calculate married rate, divorce rate, widow rate, separated rate and bachelor rate for every occupation type from the above data and create a dataframe with all the above calculated columns and occupation type.
3.	Add Work Schedule column in original dataframe containing value – (Full Time / Part Time) based on below pointers.
Full time – working hour greater than equal to 40 hours
Part time – working hour less than 40 hours
4.	Add Paid status column in original dataframe based on below condition:
a.	Income - >50K 
i.	Hour per week >= 50 then ‘Normal paid’
ii.	Hour per week < 50 then ‘Highly paid’
b.	Income - <= 50K
i.	Hour per week >= 50 then ‘Under paid’
ii.	Hour per week < 50 then ‘Normal paid’
"""


df=spark.read.csv('dbfs:/databricks-datasets/adult/adult.data',inferSchema=True)
df.display()
df.printSchema()
```

```python
from pyspark.sql.types import StructType,StructField,StringType,IntegerType,DoubleType
from pyspark.sql.functions import *

#Question 1

schma=StructType().add("Age",IntegerType(),True).add("workclass",StringType(),True).add("fnwgt",DoubleType(),True).add("education",StringType(),True)\
.add("education_num",StringType(),True).add("marital_status",StringType(),True).add("occupation",StringType(),True).add("relationship",StringType(),True).add("race",StringType(),True)\
.add("sex",StringType(),True).add("capital_gain",DoubleType(),True).add("capital_loss",DoubleType(),True).add("working_hours",DoubleType(),True)\
.add("country",StringType(),True).add("income",StringType(),True)

df=spark.read.format('csv').option('header',True).schema(schma).load('dbfs:/databricks-datasets/adult/adult.data')
df.display()
```

```python
df_result = df.select(
  df['education'], 
  # create a yes/no type col on the fly
  when( col('marital_status') == ' Never-married' , 1 ).otherwise(0).alias('is_bachelor')
).display()
```

```python
df.select("marital_status").distinct().display()
```

```python
#Question 2

df1=df.select(col('occupation'), when(col("marital_status").like("%Divo%"),1).otherwise(0).alias("Divorced rate"),
             when(col("marital_status").like("%Mar%"),1).otherwise(0).alias("Married rate"),
             when(col("marital_status").like("%Never%"),1).otherwise(0).alias("Bachelor rate"),
             when(col("marital_status").like("%Wido%"),1).otherwise(0).alias("Widowed rate"),
             when(col("marital_status").like("%Sepa%"),1).otherwise(0).alias("Separated rate"))
df1.groupBy(col("occupation")).agg(round(mean("Divorced rate"),2).alias("divorced rate"),
                                  round(mean("Married rate"),2).alias("Married rate"),
                                  round(mean("Bachelor rate"),2).alias("Bachelor rate"),
                                   round(mean("Widowed rate"),2).alias("Widowed rate"),
                                   round(mean("Separated rate"),2).alias("Separated rate"),
                                  ).sort("divorced rate","Married rate","Bachelor rate","Widowed rate","Separated rate",ascending=False).display()
```

```python
#Question 3

df.withColumn("work",when(col("working_hours")>=40,"Full Time").otherwise("Part time")).display()
```

```python
#Question 4 

df.withColumn("paid_status", when(col("income")=='>50K',when(col('working_hours')>=50,"Normal Paid").otherwise("Highly Paid"))
          .otherwise(when(col("income")=='<=50K',when(col('working_hours')>=50,"Under Paid").otherwise("normal Paid")))).display()
```
